```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
data = tf.keras.utils.image_dataset_from_directory("/content/data10/cifar10/train",image_size=(32,32),color_mode="rgb")
```

```python
model=tf.keras.Sequential([
    layers.Conv2D(9,3,input_shape=(32,32,3)),
    layers.MaxPool2D(),
    layers.Conv2D(18,3,activation="leaky_relu"),
    layers.MaxPool2D(),
    layers.Conv2D(36,3,activation="leaky_relu"),
    layers.MaxPool2D(),
    layers.Flatten(),
    layers.Dense(10),
    layers.Activation("softmax")
])
```

```python
model.summary()
```

```python
model.compile(
    loss = tf.keras.losses.categorical_crossentropy,
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)
```

```python
for features,labels in data:
  features = tf.divide(features,255)
  labels = tf.one_hot(labels,10)
  record = model.fit(
               features,
               labels,
               batch_size=50000,
               epochs=10
               )
```
